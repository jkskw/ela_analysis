```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import numpy as np

# Filtering the data for the year 2022
graduates_data_2021 = graduates_data[graduates_data['P_ROKDYP'] == 2022]

# Creating DataFrame with necessary columns
df = pd.DataFrame({
    'P_ME_ZAR': graduates_data_2021['P_ME_ZAR'],
    'P_WOJ': graduates_data_2021['P_WOJ']
})

# Adding names of wojewodztwa
df['P_WOJ_NAZWA'] = df['P_WOJ'].map(wojewodztwa_dict)

# Converting 'P_ME_ZAR' to float, replacing commas with dots
df['P_ME_ZAR'] = df['P_ME_ZAR'].astype(str).str.replace(',', '.').astype(float)

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Remove rows with NaN in 'P_ME_ZAR' and 'P_WOJ'
df.dropna(subset=['P_WOJ', 'P_ME_ZAR'], inplace=True)

# Group by wojewodztwo and calculate statistics
statystyki_P_ME_ZAR = df.groupby('P_WOJ')['P_ME_ZAR'].describe()
statystyki_P_ME_ZAR.index = statystyki_P_ME_ZAR.index.map(wojewodztwa_dict)
print(statystyki_P_ME_ZAR)

# Boxplot for zarobki distribution
plt.figure(figsize=(16, 10))
df.boxplot(column='P_ME_ZAR', by='P_WOJ_NAZWA', grid=True, showfliers=False, patch_artist=True)

# Unique colors for each wojewodztwo
num_wojewodztw = len(df['P_WOJ_NAZWA'].unique())
colors = cm.get_cmap('tab20', num_wojewodztw)
for i, patch in enumerate(plt.gca().artists):
    patch.set_facecolor(colors(i))

# Plot settings
plt.title("Rozkład średnich zarobków w województwach (bez wartości odstających)")
plt.suptitle("")
plt.xlabel("Województwo")
plt.ylabel("Średnie zarobki (P_ME_ZAR)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Bar plot for mean zarobki
plt.figure(figsize=(12, 8))
statystyki_P_ME_ZAR['mean'].plot(kind='bar', color='skyblue')
plt.title("Średnie zarobki w województwach")
plt.xlabel("Województwo")
plt.ylabel("Średnie zarobki (P_ME_ZAR)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Bar plot for median zarobki
plt.figure(figsize=(12, 8))
statystyki_P_ME_ZAR['50%'].plot(kind='bar', color='lightgreen')
plt.title("Mediana zarobków w województwach")
plt.xlabel("Województwo")
plt.ylabel("Mediana zarobków (P_ME_ZAR)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Histogram with KDE overlay for zarobki distribution
plt.figure(figsize=(14, 10))
for wojewodztwo, dane in df.groupby('P_WOJ_NAZWA'):
    sns.histplot(dane['P_ME_ZAR'], kde=True, label=wojewodztwo, bins=30, element="step", fill=True, alpha=0.3)

plt.title("Rozkład średnich zarobków (P_ME_ZAR) w województwach (Histogram z KDE)")
plt.xlabel("Średnie zarobki (P_ME_ZAR)")
plt.ylabel("Liczba obserwacji")
plt.legend(title="Województwo", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Zakładam, że masz już dwie tabele z danymi: `statystyki_P_WWB` i `statystyki_P_WWZ`
# Przykład danych dla tabel:
# statystyki_P_WWB i statystyki_P_WWZ to DataFrames z indeksami jako nazwy województw

# Utwórz DataFrame do analizy i scal dane
merged_data = pd.DataFrame({
    "Względny wskaźnik bezrobocia (P_WWB)": statystyki_P_WWB['mean'],
    "Względny wskaźnik zarobków (P_WWZ)": statystyki_P_WWZ['mean'],
    "Liczba obserwacji": statystyki_P_WWZ['count']  # Możesz użyć liczby obserwacji z obu tabel
})

# Rozmiar wykresu
plt.figure(figsize=(14, 10))

# Wykres bąbelkowy (scatter z rozmiarem markerów)
scatter = plt.scatter(
    x=merged_data["Względny wskaźnik zarobków (P_WWZ)"],
    y=merged_data["Względny wskaźnik bezrobocia (P_WWB)"],
    s=merged_data["Liczba obserwacji"] / 10,  # Skalowanie rozmiaru bąbelka, dostosuj według potrzeb
    alpha=0.6,
    c=range(len(merged_data)),  # Kolory dla każdego województwa
    cmap="tab20",  # Paleta kolorów
    edgecolor="black"
)

# Dodaj etykiety nazw województw
for i, txt in enumerate(merged_data.index):
    plt.annotate(txt, (merged_data["Względny wskaźnik zarobków (P_WWZ)"][i], 
                       merged_data["Względny wskaźnik bezrobocia (P_WWB)"][i]),
                 fontsize=10, ha='right')

# Dodaj tytuł i etykiety osi
plt.title("Względny Wskaźnik Zarobków a Względny Wskaźnik Bezrobocia w Województwach")
plt.xlabel("Względny Wskaźnik Zarobków (P_WWZ)")
plt.ylabel("Względny Wskaźnik Bezrobocia (P_WWB)")
plt.colorbar(scatter, label="Indeks województwa")
plt.grid(True)

# Wyświetl wykres
plt.show()

# Przeprowadzenie analizy korelacji dla zmiennych numerycznych
correlation_matrix = merged_data[["Względny wskaźnik bezrobocia (P_WWB)", "Względny wskaźnik zarobków (P_WWZ)"]].corr()
print("Macierz korelacji:")
print(correlation_matrix)

# Wykres korelacji
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Macierz korelacji dla względnych wskaźników")
plt.show()
```
