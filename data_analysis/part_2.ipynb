```python
def plot_graduate_distribution(data, column, study_type='1', lic_eng='eng', num_bins=10, x_range=(0, 1), title_prefix='', xlabel=''):
    """
    Plots the distribution of the number of graduates (P_N) across specified ranges of a given column.
    
    Parameters:
    data (pd.DataFrame): DataFrame containing the data to plot.
    column (str): The column for which the distribution is plotted (e.g., 'P_ME_ZAR', 'P_WWZ').
    study_type (str): The type of study ('S' for full-time, 'N' for part-time).
    num_bins (int): Number of bins for the histogram.
    x_range (tuple): Range of x-axis values for binning.
    title_prefix (str): Prefix for the plot title.
    xlabel (str): Label for the x-axis.
    """
    # Filter data for the specific study type
    data_study = data[data['P_POZIOM'] == study_type]

    if lic_eng == 'eng' and study_type == '1':
        data_study = data_study[data_study['P_N_SEMESTR'].isin([7, 8])]
    elif lic_eng == 'lic' and study_type == '1':
        data_study = data_study[~data_study['P_N_SEMESTR'].isin([7, 8])]


    # Define bin edges
    bins = np.linspace(x_range[0], x_range[1], num_bins + 1)

    # Group data by specified column bins and sum P_N
    study_groups = data_study.groupby(pd.cut(data_study[column], bins=bins))['P_N'].sum()

    # Plot the grouped data
    study_groups.plot(kind='bar', figsize=(10, 6))
    plt.title(f'{title_prefix} Distribution of Number of Graduates (P_N) Across {xlabel} Ranges for {"Full-Time" if study_type == "S" else "Part-Time"} Studies (P_FORMA = {study_type})')
    plt.xlabel(xlabel)
    plt.ylabel('Number of Graduates')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def plot_graduate_distribution(data, column, study_type='1', lic_eng='eng', num_bins=10, x_range=(0, 1), title_prefix='', xlabel='', color_scheme=('green', 'blue')):
    """
    Plots the distribution of the number of graduates (P_N) across specified ranges of a given column.
    
    Parameters:
    data (pd.DataFrame): DataFrame containing the data to plot.
    column (str): The column for which the distribution is plotted (e.g., 'P_ME_ZAR', 'P_WWZ').
    study_type (str): The type of study ('S' for full-time, 'N' for part-time).
    num_bins (int): Number of bins for the histogram.
    x_range (tuple): Range of x-axis values for binning.
    title_prefix (str): Prefix for the plot title.
    xlabel (str): Label for the x-axis.
    color_scheme (tuple): Tuple of colors for full-time and part-time studies.
    """
    # Filter data for the specific study type
    data_study = data[data['P_POZIOM'] == study_type]

    if lic_eng == 'eng' and study_type == '1':
        data_study = data_study[data_study['P_N_SEMESTR'].isin([7, 8])]
    elif lic_eng == 'lic' and study_type == '1':
        data_study = data_study[~data_study['P_N_SEMESTR'].isin([7, 8])]

    # Plot the histogram with KDE overlay
    plt.figure(figsize=(10, 6))
    sns.histplot(
        data=data_study, 
        x=column, 
        weights='P_N', 
        bins=num_bins, 
        binrange=x_range, 
        kde=False, 
        color=color_scheme[0] if study_type == 'S' else color_scheme[1], 
        edgecolor="black", 
        alpha=0.6, 
        stat="count"
    )
    plt.title(f'{title_prefix} Distribution of Number of Graduates (P_N) Across {xlabel} Ranges for {study_type}')
    plt.xlabel(xlabel)
    plt.ylabel('Number of Graduates')
    plt.xlim(x_range)
    plt.tight_layout()
    plt.show()
```

```python
def compare_graduate_salaries(graduates_data, var1, forma1, poziom1, var2, forma2, poziom2, desired_year=2018):
    # Helper function to filter and calculate median salaries
    def get_median_salaries(data, var, forma, poziom, year):
        temp_data = data[['P_ME_ZAR_DOSW_P1', 'P_ME_ZAR_DOSW_P2', 'P_ME_ZAR_DOSW_P3', 'P_ME_ZAR_DOSW_P4', 'P_ME_ZAR_DOSW_P5',
                          'P_ME_ZAR_NDOSW_P1', 'P_ME_ZAR_NDOSW_P2', 'P_ME_ZAR_NDOSW_P3', 'P_ME_ZAR_NDOSW_P4', 'P_ME_ZAR_NDOSW_P5',
                          'P_N', 'P_ROKDYP', 'P_KIERUNEK_NAZWA', 'P_FORMA', 'P_POZIOM']].copy()
        
        temp_data = temp_data[temp_data['P_ROKDYP'] == year]
        temp_data = temp_data[temp_data['P_FORMA'] == forma]
        temp_data = temp_data[temp_data['P_POZIOM'] == poziom]
        temp_data = temp_data[temp_data['P_KIERUNEK_NAZWA'].str.contains(var, case=False, na=False)]

        if temp_data.empty:
            return [0] * 5, [0] * 5  # Return zeros if no data matches

        median_experience = [
            temp_data[f'P_ME_ZAR_DOSW_P{i}'].median() for i in range(1, 6)
        ]

        median_no_experience = [
            temp_data[f'P_ME_ZAR_NDOSW_P{i}'].median() for i in range(1, 6)
        ]
        
        return median_experience, median_no_experience
    
    # Get median salaries for the first set of parameters
    med_exp1, med_no_exp1 = get_median_salaries(graduates_data, var1, forma1, poziom1, desired_year)
    # Get median salaries for the second set of parameters
    med_exp2, med_no_exp2 = get_median_salaries(graduates_data, var2, forma2, poziom2, desired_year)

    # Plot the data
    plt.figure(figsize=(14, 7))
    bar_width = 0.2  # Width of the bars
    index = np.arange(1, 6)  # Years 1 to 5

    # Plotting bars for the first comparison
    plt.bar(index - bar_width * 1.5, med_exp1, bar_width, label=f'Exp. {var1} ({forma1}{poziom1})', color='blue', edgecolor='black')
    plt.bar(index - bar_width / 2, med_no_exp1, bar_width, label=f'Non-Exp. {var1} ({forma1}{poziom1})', color='orange', edgecolor='black', alpha=0.7)
    
    # Plotting bars for the second comparison
    plt.bar(index + bar_width / 2, med_exp2, bar_width, label=f'Exp. {var2} ({forma2}{poziom2})', color='green', edgecolor='black')
    plt.bar(index + bar_width * 1.5, med_no_exp2, bar_width, label=f'Non-Exp. {var2} ({forma2}{poziom2})', color='red', edgecolor='black', alpha=0.7)

    # Adding titles and labels
    plt.title(f'Five-Year Analysis of Median Graduate Salaries for {desired_year} ("{var1}" ({forma1}{poziom1}) vs "{var2}" ({forma2}{poziom2}))')
    plt.xlabel('Years')
    plt.ylabel('Median Salary')
    plt.xticks(index, [f'Year {i}' for i in range(1, 6)])
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example call to the function
compare_graduate_salaries(graduates_data, 'zarządzanie', 'S', '2', 'informat', 'N', '2')
```
